## Access EC2 from Linux Command Line

Return to your desktop

Move the recently created .pem file to the Desktop (or some other convenient location)

Open your terminal
    - move into folder containing the .pem file
    - using cd command

Change user permissions on file
   - $ chmod 400 newkey.pem

SSH into the Amazon instance using your public DNS
    - Public DNS grabbed from EC2 dashboard
    - $ ssh -i newkey.pem ubuntu@public_dns

    - enter 'yes' when prompted to proceed

You are now accessing the command line of the EC2 instance

## Configure Jupyter Notebook

Generate configuration file for Jupyter Notebook:
    - $ jupyter notebook --generate-config

Create a folder to pass certifications:
    - $ mkdir certs
    - $ cd certs
    -  $ sudo openssl req -x509 -nodes -days 365 -newkey rsa:1024 -keyout mycert.pem -out mycert.pem

Enter whichever fields you want when prompted, they are incorporated into the certificate request

```python
Image(filename='19_create_cert.png')
```

Navigate to the configuration file:
    - $ cd ~/.jupyter
    - $ vi jupyter_notebook_config.py
            - vi will open up a text editor in the terminal

```python
Image(filename='20_access_vi.png')
```

Press 'I' in the vi editor to insert text and add the following text:
    - c = get_config()
    - c.NotebookApp.certfile = u'/home/ubuntu/certs/mycert.pem'
    - c.NotebookApp.ip = '*'
    - c.NotebookApp.open_browser = False
    - c.NotebookApp.port = 8888

Press 'ESC' to exit insert mode:
    - type :wq!
        - writes the changes and quits the editor

```python
Image(filename='21_jn_config.png')
```

Create a password to access Jupyter Notebook through the browser:
    - $ jupyter notebook password
        - enter and verify a password

Make sure jupyter notebook actually works:
    - navigate to home folder
        - $ cd
    
    - $ jupyter notebook

```python
Image(filename='22_set_jn_password.png')
```

Open Browser and navigate to:

## Install Filezilla

Use Filezilla to transfer files from your personal pc to the cloud instance of EC2

Open Desktop terminal
    - $ sudo apt-get install filezilla

```python
Image(filename='23_get_filezilla.png')
```

Open Filezilla from Applications Manager

```python
Image(filename='24_open_filezila.png')
```

Navigate to File > Site Manager on nav bar
    - Click 'New Site'
        - name it 'ec2_instance' or something else relevant
        - use your public_dns as the host
        - change protocol to SFTP
        - change logon type to Key File
            - User: ubuntu
            - Key File: click Browse -> stored .pem file
                
        - Click OK
        
    - Click on Dropdown Menu from nav bar
        - click on the newly created site 'ec2_instance'
        - click Ok

```python
Image(filename='25_connect_thru_filezilla.png')
```

Drag and Drop files between both server instances as needed
